This piece of coding demonstrates my ability to plot graphs. Two random DNA sequences are generated and compared using numpy and matplotlib. Also two short self-made functions in "dna_tools" will be used. This piece of coding has the following parts;
1) Create two random variabes with random DNA nucleotides length of 500
2) Count nucleotide usages using my self-made functions
3) Plot a line plot
4) Plot a histogram plot

```python
# (1) Random DNA nucleotides 500 length
import random
from dna_tools import get_counts
import matplotlib.pyplot as plt

random_seq1 = '' 
random_seq2 = '' 

for i in range(500): 
    random_seq1 += random.choice('ATGC') 
    random_seq2 += random.choice('ATGC') 
```

```python
 # (2) Count nucleotides
random_seq1_counts = get_counts(random_seq1,1)
random_seq2_counts = get_counts(random_seq2,1)

basepairs = list(random_seq1_counts.keys()) 
basepairs_sorted = sorted(basepairs) 

rand1_num = []  
rand2_num = [] 
for bp in basepairs_sorted: 
    rand1_num += [random_seq1_counts[bp]] 
    rand2_num += [random_seq2_counts[bp]]

print (basepairs, rand1_num, random_seq1_counts)
print (basepairs, rand2_num)
```

```python
# (3) Plot a line graph

plt.xticks([1,2,3,4],basepairs_sorted) 


plt.plot([1,2,3,4], rand1_num, marker='s', label='seq_1') 
plt.plot([1,2,3,4], rand2_num, marker='o', linestyle='--', label='seq_2')
plt.legend()
plt.show()
```

```python
 # (4) Plot a histogram 
import numpy as np

labels = basepairs_sorted
seq1_values = rand1_num
seq2_values = rand2_num

x = np.arange(len(labels))  # label locations
width = 0.45  

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, seq1_values, width, label='seq 1')
rects2 = ax.bar(x + width/2, seq2_values, width, label='seq 2')

ax.set_ylabel('Count')
ax.set_title('Base counts for two sequences')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

fig.tight_layout()

plt.show()
```
